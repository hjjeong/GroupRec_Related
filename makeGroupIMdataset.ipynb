```python

with open("data/douban/group_users.csv", "w") as fr:
    fr.write("group,user\n")
    with open("data/douban/douban.train.gu", "r") as f:
        lines = f.readlines()
        for line in lines:
            line = line.strip()
            a = line.split('\t')
            g = int(a[0])
            for m in a[1].split(','):
                if len(m)==0:
                    continue
                fr.write(str(g)+','+str(float(m))+'\n')
```

```python
with open("data/douban/train_gi.csv", "w") as fr:
    fr.write("group,item\n")
    with open("data/douban/douban.train.rating", "r") as f:
        lines = f.readlines()
        for line in lines:
            line = line.strip()
            a = line.split('\t')
            g = int(a[0])
            v = int(a[1])
            fr.write(str(g)+','+str(v)+'\n')
```

```python
group_user_dict = {}
user_item_dict = {}
with open("data/douban/train_ui.csv", "w") as fr:
    fr.write("user,item\n")
    with open("data/douban/douban.train.gu", "r") as f:
        lines = f.readlines()
        for line in lines:
            line = line.strip()
            a = line.split('\t')
            g_id = int(a[0])
            g_users = set()
            for m in a[1].split(','):
                if len(m)==0:
                    continue
                user_id = int(m)
                g_users.add(int(user_id ))
                if user_id not in user_item_dict:
                    user_item_dict[user_id] = set()
            group_user_dict[g_id] = g_users

    with open("data/douban/douban.train.rating", "r") as f:
        lines = f.readlines()
        for line in lines:
            line = line.strip()
            a = line.split('\t')
            g = int(a[0])
            v = int(a[1])
            for user in group_user_dict[g]:
                user_item_dict[user].add(v)
    for key in user_item_dict:
        for value in user_item_dict[key]:
            fr.write(str(key)+','+str(value)+'\n')
```
